```python
def buscar_exoplaneta(df, nombre_exoplaneta):
    exoplaneta = df[df['pl_name'] == nombre_exoplaneta]
    
    if not exoplaneta.empty:
        exoplaneta_confirmado = exoplaneta[exoplaneta['soltype'].str.contains('Published Confirmed', na=False)]
        
        if not exoplaneta_confirmado.empty:
            ultimo_exoplaneta = exoplaneta_confirmado.sort_values(by='rowupdate', ascending=False).iloc[0]
            sistema_planetario = exoplaneta['hostname']#.nunique()
            print(sistema_planetario)
            return ultimo_exoplaneta, sistema_planetario
    return None, None
```

```python
exoplaneta_info, sistema_planetario = buscar_exoplaneta(df, nombre_exoplaneta)

if exoplaneta_info is not None:
    print(exoplaneta_info)
    print(f'El exoplaneta {nombre_exoplaneta} pertenece al sistema planetario: {sistema_planetario}')
else:
    print(f'No se encontró el exoplaneta {nombre_exoplaneta}.')
```

```python
Gaia.MAIN_GAIA_TABLE = "gaiadr3.gaia_source"
```

```python
coord = SkyCoord(ra=ra, dec=dec, unit=(u.degree, u.degree), frame='icrs')

width = u.Quantity(20, u.deg)

height = u.Quantity(20, u.deg)

gaia_data = Gaia.query_object_async(coordinate=coord, width=width, height=height)
```

```python
def graficar_exoplaneta(exoplaneta, gaia_data):
    plt.figure(figsize=(10, 10))
    tamaño_puntos = gaia_data['phot_g_mean_flux'] * 0.0000005
    plt.scatter(gaia_data['ra'], gaia_data['dec'], 
                s=tamaño_puntos, color='blue', alpha=0.5, label='Estrellas de Gaia', marker= 'D')
    plt.scatter(exoplaneta['ra'], exoplaneta['dec'], color='red', s=1, label='Exoplaneta')

    # Etiquetas y título
    plt.title(f'Ubicación de {exoplaneta["pl_name"].values[0]}')
    plt.xlabel('RA (degrees)')
    plt.ylabel('Dec (degrees)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a la función con los datos
graficar_exoplaneta(exoplaneta, gaia_data)
```

```python
# Función para graficar el exoplaneta y las estrellas más allá del exoplaneta (mantener 2D)
def graficar_exoplaneta_filtrado(exoplaneta, gaia_data):
    # Calcular las distancias en parsecs de las estrellas usando el paralaje
    gaia_data['distancia_pc'] = 1000 / gaia_data['parallax']
    
    # Calcular la distancia del exoplaneta en parsecs
    distancia_exoplaneta = exoplaneta['sy_dist'].values[0]
    
    # Filtrar las estrellas que están más lejos que el exoplaneta
    gaia_data_filtrado = gaia_data[gaia_data['distancia_pc'] > distancia_exoplaneta]

    # Graficar solo las estrellas más lejanas que el exoplaneta
    plt.figure(figsize=(10, 10))
    tamaño_puntos = gaia_data_filtrado['phot_g_mean_flux'] * 0.000001
    plt.scatter(gaia_data_filtrado['ra'], gaia_data_filtrado['dec'], 
                s=tamaño_puntos, color='blue', alpha=0.5, label='Estrellas de Gaia', marker= 'D')

    # Graficar el exoplaneta
    plt.scatter(exoplaneta['ra'], exoplaneta['dec'], color='red', s=50, label='Exoplaneta')

    # Etiquetas y título
    plt.title(f'Ubicación de {exoplaneta["pl_name"].values[0]}')
    plt.xlabel('RA (degrees)')
    plt.ylabel('Dec (degrees)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a la función con los datos
graficar_exoplaneta_filtrado(exoplaneta, gaia_data)
```
